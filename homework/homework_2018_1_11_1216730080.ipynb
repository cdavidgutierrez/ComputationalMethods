```python
%pylab inline
#import matplotlib.pyplot as plt
import scipy.optimize as opt
#import numpy as np


# DATA ------------------------------------------------------------------------
T_values = np.array([222, 284, 308.5, 333, 358, 411, 477, 518, 880, 1080, 1259])
C_values = np.array([0.1282, 0.2308, 0.2650, 0.3120 , 0.3547, 0.4530, 0.5556, 0.6154, 0.8932, 0.9103, 0.9316])

x_samp = T_values
y_samp = C_values  
```

```python
# GENERAL EQUATION ------------------------------------------------------------
def func(x, A, c, d):
    return A*np.exp(c*x) + d
```

```python
# SURVEY ----------------------------------------------------------------------
# Plotting Sampling Data
plt.plot(x_samp, y_samp, "ko", label="Data")

x_lin = np.linspace(0, x_samp.max(), 50)                   # 50 evenly spaced digits between 0 and max

# Trials
A, c, d = -1, -1e-2, 1
y_trial1 = func(x_lin,  A,     c, d)
y_trial2 = func(x_lin, -1, -1e-3, 1)
y_trial3 = func(x_lin, -1, -3e-3, 1)

plt.plot(x_lin, y_trial1, "b--", label="Trial 1")
plt.plot(x_lin, y_trial2, "y--", label="Trial 2")
plt.plot(x_lin, y_trial3, "g--", label="Trial 3")
plt.legend()
plt.show()
```

```python
y_trial4 = func(x_lin, -1.59, -2.4e-3, 1.04)
```

```python
plt.plot(x_lin, y_trial4, "r--", label='Trial 4')
plt.plot(x_samp, y_samp, "ko", label='Data')
plt.legend()
plt.grid()
```
